## –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –º–æ–¥–µ–ª—å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤

–ü–æ—Å—Ç—Ä–æ–∏–º –º–æ–¥–µ–ª—å –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤, –≥–¥–µ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ ‚Äî —Ñ–∞–∫—Ç –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–∞ –≤ —Å–ª–µ–¥—É—é—â–µ–º –º–µ—Å—è—Ü–µ:
   - –†–∞–∑–æ–±—ä–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É —Ñ—É–Ω–∫—Ü–∏–µ–π train_test_split().
   - –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:
        - –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π,
        - —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º.
   - –û—Ü–µ–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision –∏ recall –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –°—Ä–∞–≤–Ω–∏–º –ø–æ –Ω–∏–º –º–æ–¥–µ–ª–∏. –ö–∞–∫–∞—è –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –ª—É—á—à–µ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫?

–ù–µ –∑–∞–±—É–¥–µ–º —É–∫–∞–∑–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä random_state –ø—Ä–∏ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–∏ –≤—ã–±–æ—Ä–∫–∏ –∏ –∑–∞–¥–∞–Ω–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞.


```python
y = data_m['churn']
X = data_m.drop(['churn'], axis=1) 
```

–†–∞–∑–±–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫–∏

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

–û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞  train-–≤—ã–±–æ—Ä–∫–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π

```python
#–æ–±—É—á–∏–º  ¬´—Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ç–æ—Ä¬ª –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
scaler = StandardScaler()
scaler.fit(X_train)
#–ø—Ä–∏–º–µ–Ω–∏–º ¬´—Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ç–æ—Ä¬ª –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –∏ –æ–±—É—á–∞—é—â–∏–µ–π –≤—ã–±–æ—Ä–∫–∏
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
#X_train_st = X_train
#X_test_st = X_test
```

```python
print('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –Ω–∞ test –∏ train –≤—ã–±–æ—Ä–∫–∞—Ö')
sns.kdeplot(y_train)
sns.kdeplot(y_test)
plt.show()
```

–í–∏–¥–∏–º, —á—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ —Å—Ö–æ–∂–∏.

```python
lr_model = LogisticRegression(solver='liblinear', random_state=0)
lr_model.fit(X_train_st, y_train)
```

```python
probabilities_lr = lr_model.predict_proba(X_test_st)[:, 1]
predictions_lr = lr_model.predict(X_test_st)

print('Accuracy: {:.4f}'.format(accuracy_score(y_test, predictions_lr)))
print('Precision: {:.4f}'.format(precision_score(y_test, predictions_lr)))
print('Recall: {:.4f}'.format(recall_score(y_test, predictions_lr)))
print('F1: {:.4f}'.format(f1_score(y_test, predictions_lr)))
print('ROC_AUC: {:.4f}'.format(roc_auc_score(y_test, predictions_lr)))
```

–û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞  train-–≤—ã–±–æ—Ä–∫–µ —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
rf_model.fit(X_train, y_train)
```

```python
probabilities_rf = rf_model.predict_proba(X_test)[:, 1]
predictions_rf = rf_model.predict(X_test)
print('Accuracy: {:.4f}'.format(accuracy_score(y_test, predictions_rf)))
print('Precision: {:.4f}'.format(precision_score(y_test, predictions_rf)))
print('Recall: {:.4f}'.format(recall_score(y_test, predictions_rf)))
print('F1: {:.4f}'.format(f1_score(y_test, predictions_rf)))
print('ROC_AUC: {:.4f}'.format(roc_auc_score(y_test, predictions_rf)))
```

2. –î–ª—è –≤–µ—Ä–Ω–æ—Å—Ç–∏ –¥–æ–±–∞–≤–∏–º –º–µ—Ç—Ä–∏–∫–∏ F1 –∏ ROC_AUC. –ò –ø–æ –æ–±–µ–∏–º –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É LogisticRegression
3. –ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫ –ª—É—á—à–µ —Å–µ–±—è –ø–æ–∫–∞–∑–∞–ª–∞ –º–æ–¥–µ–ª—å LogisticRegression.

## –°–¥–µ–ª–∞—Ç—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é –∫–ª–∏–µ–Ω—Ç–æ–≤

### –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ

```python
scaler = StandardScaler() # —Å–æ–∑–¥–∞—ë–º –æ–±—ä–µ–∫—Ç –∫–ª–∞—Å—Å–∞ scaler (–Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ç–æ—Ä)
x_sc = scaler.fit_transform(data_m.drop(columns = ['churn'])) # –æ–±—É—á–∞–µ–º –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ç–æ—Ä –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö 
print(x_sc[:10]) 
```

### –ü–æ—Å—Ç—Ä–æ–∏–º –º–∞—Ç—Ä–∏—Ü—É —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–µ–π linkage() –Ω–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—É. –í–Ω–∏–º–∞–Ω–∏–µ: –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—ã –º–æ–∂–µ—Ç –∑–∞–Ω—è—Ç—å –≤—Ä–µ–º—è! –ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç–µ, –∫–∞–∫–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –º–æ–∂–Ω–æ –≤—ã–¥–µ–ª–∏—Ç—å.

```python
linked = linkage(x_sc, method = 'ward') 
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('–î–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º–∞ –¥–ª—è —Ñ–∏—Ç–Ω–µ—Å—Å –∫–ª—É–±–∞')
plt.show()
```

```python
km = KMeans(n_clusters=5, random_state=0) # –∑–∞–¥–∞—ë–º —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, —Ä–∞–≤–Ω–æ–µ 5, –∏ —Ñ–∏–∫—Å–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ random_state –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
labels = km.fit_predict(x_sc)
data_m['cluster'] = labels
data_m.info()
data_m.head()
```

### –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä–æ–≤. –ú–æ–∂–Ω–æ –ª–∏ —Å—Ä–∞–∑—É —á—Ç–æ-—Ç–æ –∑–∞–º–µ—Ç–∏—Ç—å?
